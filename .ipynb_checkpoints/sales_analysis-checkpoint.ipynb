## Categorical Columns Overview

This section visualizes the key categorical columns in the dataset to understand their distribution:  
- Product `Category`  
- Product `Sub-Category`  
- `Region`


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

fig, axes = plt.subplots(1, 3, figsize=(20,6))
plt.subplots_adjust(wspace=0.4)

# 1. Category distribution
cat_counts = df['Category'].value_counts()
sns.barplot(x=cat_counts.index, y=cat_counts.values, palette='Blues_r', ax=axes[0])
axes[0].set_title('Category Distribution', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Count', fontsize=12)
axes[0].set_xlabel('')
for i, v in enumerate(cat_counts.values):
    axes[0].text(i, v + max(cat_counts.values)*0.01, str(v), ha='center', fontweight='bold')

# 2. Sub-Category distribution
subcat_counts = df['Sub-Category'].value_counts()
sns.barplot(x=subcat_counts.values, y=subcat_counts.index, palette='Greens_r', ax=axes[1])
axes[1].set_title('Sub-Category Distribution', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Count', fontsize=12)
axes[1].set_ylabel('')
for i, v in enumerate(subcat_counts.values):
    axes[1].text(v + max(subcat_counts.values)*0.01, i, str(v), va='center', fontweight='bold')

# 3. Region distribution
region_counts = df['Region'].value_counts()
sns.barplot(x=region_counts.index, y=region_counts.values, palette='Oranges_r', ax=axes[2])
axes[2].set_title('Region Distribution', fontsize=14, fontweight='bold')
axes[2].set_ylabel('Count', fontsize=12)
axes[2].set_xlabel('')
for i, v in enumerate(region_counts.values):
    axes[2].text(i, v + max(region_counts.values)*0.01, str(v), ha='center', fontweight='bold')

plt.show()
```

# 5. Data Cleaning & Feature Engineering

Prepare dataset for analysis by:

1. Converting dates to datetime
2. Cleaning Postal Codes
3. Creating new temporal features: Year, Month, Weekday



```python
# Convert Order Date and Ship Date to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Clean Postal Code (remove '.0' and ensure string type)
df['Postal Code'] = df['Postal Code'].astype(str).str.replace(r'\.0$', '', regex=True)

# Create temporal features for analysis
df['Order_Year'] = df['Order Date'].dt.year
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Weekday'] = df['Order Date'].dt.day_name()

# Display first 5 rows after cleaning and feature engineering
df[['Order Date','Ship Date','Postal Code','Order_Year','Order_Month','Order_Weekday']].head()
```

# 6.Key Business Metrics

This section calculates and visualizes key business metrics including:  

- **Total Sales**  
- **Total Profit**  
- **Average Order Value**  
- **Unique Customers**


```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# Metrics calculation
total_sales = df['Sales'].sum()
total_profit = df['Profit'].sum()
avg_order_value = df['Sales'].mean()
unique_customers = df['Customer ID'].nunique()

# Print metrics
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Average Order Value: ${avg_order_value:,.2f}")
print(f"Unique Customers: {unique_customers}")

# Visualization
metrics = ['Total Sales', 'Total Profit', 'Avg Order Value', 'Unique Customers']
values = [total_sales, total_profit, avg_order_value, unique_customers]

plt.figure(figsize=(8,5))
bars = sns.barplot(x=metrics, y=values, palette='Set2')
plt.title("Key Business Metrics", fontsize=14, fontweight='bold')
plt.ylabel("Value")
plt.xticks(rotation=15)

# Add value labels on bars
for bar in bars.patches:
    height = bar.get_height()
    bars.annotate(f'{height:,.0f}', 
                  xy=(bar.get_x() + bar.get_width() / 2, height),
                  xytext=(0,5),  # 5 points vertical offset
                  textcoords="offset points",
                  ha='center', va='bottom', fontsize=10)

plt.show()
```

1. Total Sales & Profit
2. Most profitable Category
3. Top 10 States by Sales
4. Scatter plot: Discount vs Profit

## 10.Temporal Analysis

Analyze sales and profit trends over time to identify seasonality, growth, and performance patterns.

- Monthly Sales Trend
- Monthly Profit Trend
- Sales vs Profit (overlay)




```python
# Overlay Chart: Sales, Profit & AOV (Smoothed)

import matplotlib.pyplot as plt

# Monthly aggregation
monthly = df.groupby(pd.Grouper(key="Order Date", freq="M")).agg(
    {"Sales":"sum","Profit":"sum","Order ID":"nunique"}
).reset_index()

# Calculate Average Order Value
monthly["AOV"] = monthly["Sales"] / monthly["Order ID"]

# Add 3-month rolling averages
monthly["Sales_Smooth"] = monthly["Sales"].rolling(3).mean()
monthly["Profit_Smooth"] = monthly["Profit"].rolling(3).mean()
monthly["AOV_Smooth"] = monthly["AOV"].rolling(3).mean()

# Plot overlay
plt.figure(figsize=(12,6))
plt.plot(monthly["Order Date"], monthly["Sales_Smooth"], color="steelblue", linewidth=2, label="Sales (3M Trend)")
plt.plot(monthly["Order Date"], monthly["Profit_Smooth"], color="seagreen", linewidth=2, label="Profit (3M Trend)")
plt.plot(monthly["Order Date"], monthly["AOV_Smooth"], color="darkorange", linewidth=2, label="AOV (3M Trend)")

plt.title("Monthly Sales, Profit & AOV Trends (Smoothed)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Amount")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# 11.Top Products Analysis

- Top 10 Loss-Making Products
- Top 10 Profit-Making Products

## Top Loss-Making Products
This section identifies products generating the highest losses.  
Highlighting these products helps in making decisions regarding pricing adjustments, promotional strategies, or potential discontinuation.


```python
# Identify top loss-making products
loss_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10)
print("=== Top Loss-Making Products ===")
print(loss_products)

# Visualization
plt.figure(figsize=(10,5))
loss_products.plot(kind='barh', color='salmon', title="Top 10 Loss-Making Products")
plt.xlabel("Profit (Negative Values)")
plt.ylabel("Product Name")
plt.show()
```

## Top Profit-Making Products
This analysis highlights the top 10 products that generate the highest profit for the business.  
By identifying these products, businesses can:  
- Focus on **boosting sales** of high-performing products.  
- **Maintain sufficient inventory** to avoid stock-outs.  
- Design **promotional campaigns** around profitable items.  

The visualization displays the top 10 products ranked by their total profit contribution, allowing management to make **data-driven decisions** on which products should receive more attention.  


```python
# Identify top 10 profit-making products
profit_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
bars = plt.barh(profit_products.index, profit_products.values, color="lightgreen")

# Highlight most profitable product
bars[-1].set_color("green")

# Add value labels
for bar in bars:
    plt.text(bar.get_width() - (0.05*abs(max(profit_products.values))), 
             bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():,.0f}',
             va='center', ha='right', fontsize=10, color="black")

plt.title("Top 10 Profit-Making Products", fontsize=14)
plt.xlabel("Total Profit")
plt.ylabel("Product Name")
plt.tight_layout()
plt.show()
```

## Profit vs Loss Products Dashboard
This dashboard provides a **side-by-side comparison** of the top 10 profit-making products and the top 10 loss-making products.  

- **Left Panel (Green)** → Top 10 Profit-Making Products  
  - Highlights the items contributing the most to overall profitability.  
  - Helps management prioritize high-performing products.  

- **Right Panel (Red)** → Top 10 Loss-Making Products  
  - Highlights the items causing maximum financial losses.  
  - Useful for deciding whether to **adjust pricing**, **reduce discounts**, or even **discontinue products**.  

This combined view enables stakeholders to **compare both positive and negative outliers simultaneously**, ensuring better **strategic decision-making** regarding product portfolio management.  


```python
# Top 10 Profit & Loss Products
profit_products = df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False).head(10)
loss_products = df.groupby('Product Name')['Profit'].sum().sort_values().head(10)

# Plot side-by-side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Left: Top Profit Products
bars1 = axes[0].barh(profit_products.index, profit_products.values, color="lightgreen")
bars1[-1].set_color("green")  # Highlight most profitable
axes[0].set_title("Top 10 Profit-Making Products", fontsize=14)
axes[0].set_xlabel("Total Profit")
for bar in bars1:
    axes[0].text(bar.get_width() - (0.05*abs(max(profit_products.values))),
                 bar.get_y() + bar.get_height()/2,
                 f'{bar.get_width():,.0f}', va='center', ha='right', fontsize=9, color="black")

# Right: Top Loss Products
bars2 = axes[1].barh(loss_products.index, loss_products.values, color="salmon")
bars2[0].set_color("red")  # Highlight worst loss
axes[1].set_title("Top 10 Loss-Making Products", fontsize=14)
axes[1].set_xlabel("Total Profit (Negative Values)")
for bar in bars2:
    axes[1].text(bar.get_width() + (0.01*abs(min(loss_products.values))),
                 bar.get_y() + bar.get_height()/2,
                 f'{bar.get_width():,.0f}', va='center', ha='left', fontsize=9, color="black")

plt.tight_layout()
plt.show()
```

# 12. Discount Impact Analysis

1. Scatter plot: Discount vs Profit
2. Regression line and correlation
3. Sales vs Profit colored by Discount

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set(style="whitegrid")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18,6))

# --- Left: Discount vs Profit with regression ---
corr = df['Discount'].corr(df['Profit'])
sns.regplot(
    x='Discount', y='Profit', data=df,
    scatter_kws={'s':70, 'alpha':0.7, 'color':'#1f77b4', 'edgecolor':'k'},
    line_kws={'color':'#d62728', 'linewidth':2},
    ax=axes[0]
)
axes[0].set_title('Discount vs Profit', fontsize=16, weight='bold')
axes[0].set_xlabel('Discount', fontsize=12)
axes[0].set_ylabel('Profit', fontsize=12)
# Annotate correlation
axes[0].text(0.05, 0.95, f'Correlation: {corr:.2f}', transform=axes[0].transAxes,
             fontsize=12, weight='bold', verticalalignment='top', 
             bbox=dict(boxstyle="round,pad=0.3", fc="yellow", alpha=0.3))

# --- Right: Sales vs Profit colored by Discount ---
scatter = axes[1].scatter(
    df['Sales'], df['Profit'], c=df['Discount'], cmap='viridis', s=70, alpha=0.8, edgecolor='k'
)
axes[1].set_title('Sales vs Profit colored by Discount', fontsize=16, weight='bold')
axes[1].set_xlabel('Sales', fontsize=12)
axes[1].set_ylabel('Profit', fontsize=12)
cbar = fig.colorbar(scatter, ax=axes[1])
cbar.set_label('Discount', fontsize=12, weight='bold')

plt.tight_layout()
plt.show()
```

```python
#  Final Summary Dashboard (Improved)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(18,14))

# Global title
plt.suptitle("Final Summary Dashboard", fontsize=18, fontweight='bold')

# 1. Sales by Category
plt.subplot(3,2,1)
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=cat_sales.index, y=cat_sales.values, palette="Set2")
plt.title("Sales by Category", fontsize=12, fontweight='bold')
plt.ylabel("Sales")
plt.xlabel("Category")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# 2. Profit by Category
plt.subplot(3,2,2)
cat_profit = df.groupby('Category')['Profit'].sum().sort_values(ascending=False)
sns.barplot(x=cat_profit.index, y=cat_profit.values, palette="Set2")
plt.title("Profit by Category", fontsize=12, fontweight='bold')
plt.ylabel("Profit")
plt.xlabel("Category")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# 3. Sales by Region
plt.subplot(3,2,3)
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=region_sales.index, y=region_sales.values, palette="Set2")
plt.title("Sales by Region", fontsize=12, fontweight='bold')
plt.ylabel("Sales")
plt.xlabel("Region")
plt.grid(axis="y", linestyle="--", alpha=0.6)

# 4. Discount vs Profit (Enhanced)
plt.subplot(3,2,4)
sns.scatterplot(
    x='Discount', y='Profit', data=df,
    size='Sales', hue='Sales',
    alpha=0.6, palette='coolwarm', legend=False
)
plt.title("Discount vs Profit (Size ~ Sales)", fontsize=12, fontweight='bold')
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.grid(True, linestyle="--", alpha=0.6)

# 5. Monthly Sales & Profit Trend (Overlay)
plt.subplot(3,1,3)
monthly_data = df.groupby(df['Order Date'].dt.to_period("M"))[['Sales','Profit']].sum()
monthly_data.index = monthly_data.index.to_timestamp()

plt.plot(monthly_data.index, monthly_data['Sales'], marker='o', label='Sales', color='steelblue')
plt.plot(monthly_data.index, monthly_data['Profit'], marker='o', label='Profit', color='seagreen')

plt.title("Monthly Sales & Profit Trend", fontsize=12, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Amount")
plt.xticks(rotation=30)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)

plt.tight_layout(rect=[0,0,1,0.96])
plt.show()
```
